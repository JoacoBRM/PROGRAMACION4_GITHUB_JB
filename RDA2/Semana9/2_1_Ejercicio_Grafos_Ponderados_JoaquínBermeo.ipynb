## Ejercicio: Grafo de rutas entre ciudades

### Enunciado

Una empresa de transporte desea modelar un sistema de rutas entre ciudades para estimar los costos de viaje. El grafo se compone de las siguientes ciudades:

- Quito
- Ambato
- Riobamba
- Cuenca

Y las rutas con sus respectivos costos (en USD):

- Quito → Ambato: 10
- Quito → Riobamba: 18
- Ambato → Cuenca: 25
- Riobamba → Cuenca: 20

### Objetivo del estudiante

1. Crear un grafo dirigido y ponderado con `networkx`.
2. Representar el grafo visualmente.
3. Mostrar los pesos de las rutas.
4. Determinar el **camino más barato** entre Quito y Cuenca usando un algoritmo como `shortest_path` o `dijkstra_path`.

```python
# Importa la librería NetworkX para trabajar con grafos y Matplotlib para graficar
import networkx as nx
import matplotlib.pyplot as plt

# Crea un grafo dirigido (las aristas tienen dirección)
G = nx.DiGraph()

# Define una lista de ciudades que serán los nodos del grafo
vertices = ['Quito', 'Ambato', 'Riobamba', 'Cuenca']

# Agrega los nodos al grafo
G.add_nodes_from(vertices)

# Agrega las aristas al grafo, cada una con un peso (distancia entre ciudades)
G.add_weighted_edges_from([
    ('Quito', 'Ambato', 10),     # De Quito a Ambato con peso 10
    ('Quito', 'Riobamba', 18),   # De Quito a Riobamba con peso 18
    ('Ambato', 'Cuenca', 25),    # De Ambato a Cuenca con peso 25
    ('Riobamba', 'Cuenca', 20)   # De Riobamba a Cuenca con peso 20
])

# Calcula una disposición visual para los nodos del grafo
pos = nx.spring_layout(G)

# Dibuja el grafo con etiquetas, color y tamaño personalizados para los nodos
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000)

# Dibuja las etiquetas de los pesos en las aristas
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'weight'))

# Muestra el grafo en una ventana gráfica
plt.title('Grafo de rutas entre ciudades')
plt.show()

```

```python
# Calcula el camino más corto desde 'Quito' hasta 'Cuenca' usando el algoritmo de Dijkstra
camino = nx.dijkstra_path(G, source="Quito", target="Cuenca")

# Calcula el costo total (suma de los pesos) del camino más corto
costo_total = nx.dijkstra_path_length(G, source="Quito", target="Cuenca")

# Muestra el camino más corto encontrado
print("Camino más corto:", camino)

# Muestra el costo total del camino más corto
print("Costo total:", costo_total)
```

# **Contexto**
Este grafo puede ser utilizado en situaciones reales donde se necesite planificar un viaje entre ciudades, como por ejemplo, desde Quito hasta Cuenca, y se desee encontrar la ruta más corta o económica.Cada nodo representa una ciudad, y cada conexión (arista) tiene un peso que indica la distancia entre ellas. Usando este modelo, se puede aplicar el algoritmo de Dijkstra para determinar el camino más eficiente, lo cual puede ayudar a:

1. Ahorrar gasolina al recorrer menos kilómetros.
2. Reducir el tiempo de viaje al evitar rutas más largas.
